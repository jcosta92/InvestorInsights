```python
main_ticker="AAPL"
```

```python
# =========================
#  Imports
# =========================
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np


# =========================
#  MarketBeat – Competitors
# =========================

def get_competitors_from_marketbeat(ticker: str, verbose: bool = True) -> pd.DataFrame:
    """
    1) Takes a ticker (e.g. 'AAPL')
    2) Tries multiple MarketBeat exchange URLs until one works
    3) Scrapes that working page for competitors
    4) Returns a DataFrame with a single column 'ticker',
       first row is the main ticker.
    """

    ticker = ticker.upper()

    # Exchanges we will brute-force in this order
    exchanges = ["NASDAQ", "NYSE", "ARCA", "AMEX", "OTC"]

    headers = {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/120.0.0.0 Safari/537.36"
        )
    }

    def _extract_from_paragraph(soup: BeautifulSoup):
        """
        Tries to extract tickers from the sentence:
        'The main competitors of X include ... (TICKER) ...'
        """
        para = soup.find(
            string=lambda s: s
            and "The main competitors of" in s
            and "(" in s
        )
        if not para:
            return []

        text = para.strip()
        peers = re.findall(r"\(([A-Za-z\.]+)\)", text)
        return [p.upper() for p in peers]

    def _extract_from_table(soup: BeautifulSoup):
        """
        Fallback: uses the 'XYZ Competitors List' table.
        Extracts tickers from the first column of that table.
        """
        heading = soup.find(string=lambda s: s and "Competitors List" in s)
        if not heading:
            return []

        heading_tag = heading.parent
        table = heading_tag.find_next("table")
        if not table:
            return []

        tickers = []
        rows = table.find_all("tr")
        for tr in rows[1:]:  # skip header
            tds = tr.find_all("td")
            if not tds:
                continue
            # example cell text: "AAPL Apple Inc."
            first_cell_text = tds[0].get_text(" ", strip=True)
            m = re.match(r"([A-Za-z\.]+)", first_cell_text)
            if not m:
                continue
            t = m.group(1).upper()
            tickers.append(t)
        return tickers

    # Try each exchange until one gives us competitors
    for ex in exchanges:
        url = (
            f"https://www.marketbeat.com/stocks/{ex}/{ticker}/"
            f"competitors-and-alternatives/"
        )

        if verbose:
            print(f"Trying {ex}: {url}")

        try:
            resp = requests.get(url, headers=headers, timeout=15)
        except requests.RequestException as e:
            if verbose:
                print(f"  ⚠ Request error for {ex}: {e}")
            continue

        if resp.status_code != 200:
            if verbose:
                print(f"  ⚠ Status code {resp.status_code} for {ex}, skipping.")
            continue

        soup = BeautifulSoup(resp.text, "html.parser")

        # 1) Try paragraph method
        peers = _extract_from_paragraph(soup)

        # 2) If paragraph not found, try competitors table
        if not peers:
            peers = _extract_from_table(soup)

        if not peers:
            if verbose:
                print(f"  ⚠ No competitors found on {ex}, trying next exchange.")
            continue

        # Build final tickers list: main ticker first, then unique peers
        all_tickers = [ticker] + [p for p in peers if p != ticker]

        if verbose:
            print(f"  ✔ Using exchange {ex}, found peers:", all_tickers)

        return pd.DataFrame({"ticker": all_tickers})

    # If nothing worked, just return the main ticker
    if verbose:
        print("❌ No competitors found on any tested exchange. Returning only main ticker.")
    return pd.DataFrame({"ticker": [ticker]})


# =========================
#  StockAnalysis – Ratios
# =========================

headers_req = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0.0.0 Safari/537.36"
    )
}


def _parse_float_sa(text: str):
    """Parse StockAnalysis numeric cell to float (handles %, commas, '--', 'n/a')."""
    if text is None:
        return np.nan
    txt = text.strip()
    if txt == "" or txt.lower() in ("n/a", "na", "--", "-"):
        return np.nan
    # remove thousands separators and percent sign
    txt = txt.replace(",", "").replace("%", "")
    try:
        return float(txt)
    except ValueError:
        return np.nan


def _get_ratios_table_for_ticker(ticker: str) -> pd.DataFrame | None:
    """
    Downloads https://stockanalysis.com/stocks/{ticker}/financials/ratios/
    and returns a DataFrame where:
      - index = metric names (e.g. 'PE Ratio', 'Return on Equity (ROE)')
      - columns = ['Current', 'FY24', 'FY23', ...] (if present)
    """
    ticker_l = ticker.lower()
    url = f"https://stockanalysis.com/stocks/{ticker_l}/financials/ratios/"
    print(f"Fetching ratios page for {ticker}: {url}")

    resp = requests.get(url, headers=headers_req, timeout=30)
    if resp.status_code != 200:
        print(f"⚠ HTTP {resp.status_code} for {ticker}.")
        return None

    soup = BeautifulSoup(resp.text, "html.parser")

    # --- Find header row containing "Fiscal Year" ---
    head_tr = None
    for tr in soup.find_all("tr"):
        ths = tr.find_all("th")
        if not ths:
            continue
        texts = [th.get_text(strip=True) for th in ths]
        if any("Fiscal Year" in t for t in texts):
            head_tr = tr
            break

    if head_tr is None:
        print(f"⚠ Could not find header row on ratios page for {ticker}.")
        return None

    # Build column headers
    cols = []
    for th in head_tr.find_all("th"):
        text = th.get_text(strip=True)
        if text in ("Fiscal Year", "2015 - 2019", "2016 - 2019"):
            continue

        # Normalize "FY 2024" -> "FY24"
        m = re.match(r"FY\s*(\d{4})", text, flags=re.I)
        if m:
            yr = m.group(1)
            cols.append(f"FY{yr[-2:]}")  # FY24 style
        else:
            # "Current", "TTM", maybe others
            cols.append(text)

    headers_full = cols

    # --- Parse all metric rows ---
    page_rows = []
    idx_names = []

    for tr in soup.find_all("tr"):
        tds = tr.find_all("td")
        if not tds:
            continue

        # Metric name from first td (prefer link text, else text)
        a = tds[0].find("a")
        if a and a.get_text(strip=True):
            metric_name = a.get_text(strip=True)
        else:
            label = tds[0].get_text(" ", strip=True)
            metric_name = re.sub(r"\s+", " ", label).strip()

        # Values in remaining tds, ignore 'Upgrade'
        vals = []
        for td in tds[1:]:
            txt = td.get_text(strip=True)
            if not txt or txt.lower().startswith("upgrade"):
                continue
            vals.append(txt)

        vals = vals[:len(headers_full)]
        if len(vals) < len(headers_full):
            vals += [None] * (len(headers_full) - len(vals))

        idx_names.append(metric_name)
        page_rows.append(vals)

    if not page_rows:
        print(f"⚠ No metric rows found on ratios page for {ticker}.")
        return None

    df = pd.DataFrame(page_rows, columns=headers_full, index=idx_names)

    # --- Unify "Current" and "TTM" into a single first column named "Current" ---
    cols_set = df.columns.tolist()
    has_current = "Current" in cols_set
    has_ttm = "TTM" in cols_set

    if has_current and has_ttm:
        combined = df["Current"].where(df["Current"].notna(), df["TTM"])
        df = df.drop(columns=["TTM"])
        df.insert(0, "Current", combined)
    elif (not has_current) and has_ttm:
        df = df.rename(columns={"TTM": "Current"})
        new_cols = ["Current"] + [c for c in df.columns if c != "Current"]
        df = df[new_cols]
    else:
        if has_current:
            new_cols = ["Current"] + [c for c in df.columns if c != "Current"]
            df = df[new_cols]

    return df


def get_stockanalysis_ratios_current(ticker: str) -> dict:
    """
    Returns a dict with:
        {
          'ticker': ...,
          'PE': float or NaN,
          'PS': ...,
          'PB': ...,
          'PFCF': ...,
          'ROA': ...,
          'ROE': ...,
          'ROIC': ...,
          'D/E': ...
        }
    Values are taken from the 'Current' column in the ratios page.
    """
    df = _get_ratios_table_for_ticker(ticker)
    result = {
        "ticker": ticker.upper(),
        "PE": np.nan,
        "PS": np.nan,
        "PB": np.nan,
        "PFCF": np.nan,
        "ROA": np.nan,
        "ROE": np.nan,
        "ROIC": np.nan,
        "D/E": np.nan,
    }

    if df is None or "Current" not in df.columns:
        return result

    # Convert Current column to float
    current_series = df["Current"].apply(
        lambda x: _parse_float_sa(x) if isinstance(x, str) else (x if x is not None else np.nan)
    )

    # Aliases to be robust to slight naming changes
    aliases = {
        "PE":   ["PE Ratio"],
        "PS":   ["PS Ratio"],
        "PB":   ["PB Ratio"],
        "PFCF": ["P/FCF Ratio"],
        "ROA":  ["Return on Assets (ROA)", "Return on Assets"],
        "ROE":  ["Return on Equity (ROE)", "Return on Equity"],
        "ROIC": [
            "Return on Capital (ROIC)",
            "Return on Invested Capital (ROIC)",
            "Return on Invested Capital"
        ],
        "D/E":  ["Debt / Equity Ratio", "Debt / Equity"],
    }

    index_names = list(df.index)

    for key, names in aliases.items():
        row_name = None
        for candidate in names:
            if candidate in index_names:
                row_name = candidate
                break
        if row_name is not None:
            val = current_series.get(row_name, np.nan)
            result[key] = val

    return result


def get_ratios_for_peers_from_stockanalysis(df_peers: pd.DataFrame) -> pd.DataFrame:
    """
    Receives a DataFrame with a 'ticker' column,
    returns a DataFrame with columns:
      ['ticker', 'PE', 'PS', 'PB', 'PFCF', 'ROA', 'ROE', 'ROIC', 'D/E']
    """
    rows = []
    for t in df_peers["ticker"]:
        try:
            ratios = get_stockanalysis_ratios_current(t)
        except Exception as e:
            print(f"⚠ Error while scraping {t}: {e}")
            ratios = {
                "ticker": t.upper(),
                "PE": np.nan,
                "PS": np.nan,
                "PB": np.nan,
                "PFCF": np.nan,
                "ROA": np.nan,
                "ROE": np.nan,
                "ROIC": np.nan,
                "D/E": np.nan,
            }
        rows.append(ratios)

    df_out = pd.DataFrame(rows)
    cols = ["ticker", "PE", "PS", "PB", "PFCF", "ROA", "ROE", "ROIC", "D/E"]
    return df_out[cols]


# =========================
#  Example end-to-end usage
# =========================

# 1) Get competitors from MarketBeat
df_peers = get_competitors_from_marketbeat(main_ticker)
print("\nPeers from MarketBeat:")
print(df_peers)

# 2) Get ratios for those tickers from StockAnalysis
df_ratios = get_ratios_for_peers_from_stockanalysis(df_peers)
print("\nRatios from StockAnalysis (Current):")
df_ratios
```

```python
import numpy as np
import pandas as pd
import yfinance as yf
import datetime as dt


def _compute_cagr(price_start: float, price_end: float, years: float) -> float:
    """CAGR em forma decimal (0.20 = 20%)."""
    if price_start <= 0 or price_end <= 0 or years <= 0:
        return np.nan
    return (price_end / price_start) ** (1.0 / years) - 1.0


def get_cagrs_from_yahoo(ticker: str) -> dict:
    """
    Calcula 5Y e 10Y price CAGR usando dados de 'Adj Close' do Yahoo Finance.

    Devolve:
        {
          '5YCAGR': valor_em_percent,  # ex: 18.5 significa 18.5% ao ano
          '10YCAGR': valor_em_percent
        }
    """
    t = ticker.upper()
    print(f"Fetching history from Yahoo for {t}...")

    # histórico máximo para garantir pelo menos 10 anos se existirem
    data = yf.Ticker(t).history(period="max", auto_adjust=False)

    if data.empty or "Adj Close" not in data.columns:
        print(f"⚠ No history or Adj Close for {t}")
        return {"5YCAGR": np.nan, "10YCAGR": np.nan}

    # garantir ordenado por data
    hist = data.sort_index()
    end_date = hist.index.max()
    price_end = float(hist.loc[end_date, "Adj Close"])

    def _cagr_for_years(years: int) -> float:
        target_start = end_date - pd.DateOffset(years=years)
        # filtrar até end_date
        sub = hist[hist.index <= end_date]

        # escolher a primeira data >= target_start; se não houver, usar a mais antiga
        cand = sub[sub.index >= target_start]
        if not cand.empty:
            start_date = cand.index[0]
        else:
            # não há histórico suficiente, usar earliest e ajustar anos reais
            start_date = sub.index[0]

        price_start = float(sub.loc[start_date, "Adj Close"])
        actual_years = (end_date - start_date).days / 365.25

        # se o período real for muito curto, devolve NaN
        if actual_years < years * 0.8:  # por ex., <4 anos para 5Y
            return np.nan

        cagr_dec = _compute_cagr(price_start, price_end, actual_years)
        return cagr_dec * 100.0  # percentagem

    cagr_5 = _cagr_for_years(5)
    cagr_10 = _cagr_for_years(10)

    return {"5YCAGR": cagr_5, "10YCAGR": cagr_10}
```

```python
# adicionar colunas vazias
df_ratios["5YCAGR"] = np.nan
df_ratios["10YCAGR"] = np.nan

for tk in df_ratios["ticker"]:
    cagrs = get_cagrs_from_yahoo(tk)
    df_ratios.loc[df_ratios["ticker"] == tk, "5YCAGR"] = cagrs["5YCAGR"]
    df_ratios.loc[df_ratios["ticker"] == tk, "10YCAGR"] = cagrs["10YCAGR"]

df_ratios
```
