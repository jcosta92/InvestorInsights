```python
import pandas as pd
import numpy as np

url = "https://stockanalysis.com/ipos/"

# Read all tables on the page
tables = pd.read_html(url)

# The IPO table is the first one found
df = tables[0]

df
```

```python
# Rename column
df = df.rename(columns={'Return': 'Return (%)'})

# Remove % and convert to numeric (invalid ones become NaN)
df['Return (%)'] = df['Return (%)'].str.replace('%', '', regex=False)
df['Return (%)'] = pd.to_numeric(df['Return (%)'], errors='coerce')
```

```python
average_return = df['Return (%)'].mean()

print("Average Return (%):", round(float(average_return), 2))
```

```python
positives = (df['Return (%)'] > 0).sum()/200*100
negatives = (df['Return (%)'] < 0).sum()/200*100
zero = (df['Return (%)'] == 0).sum()/200*100   # optional

print("Positive IPOs [%]:", positives)
print("Negative IPOs [%]:", negatives)
print("Zero return IPOs [%]:", zero)
```

```python
POS_IPOs=df[df['Return (%)'] > 0].reset_index(0)
NEG_IPOs=df[df['Return (%)'] > 0].reset_index(0)
```

```python
POS_IPOs
```

```python
average_return = POS_IPOs['Return (%)'].mean()

print("Average Return on Profitable IPOs (%):", round(float(average_return), 2))
```

```python
df['IPO Date'] = pd.to_datetime(df['IPO Date'], format='%b %d, %Y')

from datetime import datetime, timedelta

today = pd.Timestamp.today()

one_month = df[df['IPO Date'] >= today - pd.DateOffset(months=1)]
three_months = df[df['IPO Date'] >= today - pd.DateOffset(months=3)]
six_months = df[df['IPO Date'] >= today - pd.DateOffset(months=6)]
```

```python
def summarize(data):
    avg_all = data['Return (%)'].mean()

    profitable = data[data['Return (%)'] > 0]
    pct_profitable = len(profitable) / len(data) * 100 if len(data) > 0 else 0
    avg_profitable = profitable['Return (%)'].mean() if len(profitable) > 0 else 0
    
    return avg_all, pct_profitable, avg_profitable
```

```python
all_stats=summarize(df)
one_stats = summarize(one_month)
three_stats = summarize(three_months)
six_stats = summarize(six_months)

print("---- All ----")
print("Avg Return (All):", round(all_stats[0], 2), "%")
print("% Profitable:", round(all_stats[1], 2), "%")
print("Avg Return (Profitable Only):", round(all_stats[2], 2), "%")

print("---- 1 Month ----")
print("Avg Return (All):", round(one_stats[0], 2), "%")
print("% Profitable:", round(one_stats[1], 2), "%")
print("Avg Return (Profitable Only):", round(one_stats[2], 2), "%")

print("\n---- 3 Months ----")
print("Avg Return (All):", round(three_stats[0], 2), "%")
print("% Profitable:", round(three_stats[1], 2), "%")
print("Avg Return (Profitable Only):", round(three_stats[2], 2), "%")

print("\n---- 6 Months ----")
print("Avg Return (All):", round(six_stats[0], 2), "%")
print("% Profitable:", round(six_stats[1], 2), "%")
print("Avg Return (Profitable Only):", round(six_stats[2], 2), "%")
```

```python
def summarize_neg(data):
    avg_all = data['Return (%)'].mean()

    nonprofitable = data[data['Return (%)'] < 0]
    pct_nonprofitable = len(nonprofitable) / len(data) * 100 if len(data) > 0 else 0
    avg_nonprofitable = nonprofitable['Return (%)'].mean() if len(nonprofitable) > 0 else 0
    
    return avg_all, pct_nonprofitable, avg_nonprofitable
```

```python
all_stats=summarize_neg(df)
one_stats = summarize_neg(one_month)
three_stats = summarize_neg(three_months)
six_stats = summarize_neg(six_months)

print("---- All ----")
print("Avg Return (All):", round(all_stats[0], 2), "%")
print("% NonProfitable:", round(all_stats[1], 2), "%")
print("Avg Loss (NonProfitable Only):", round(all_stats[2], 2), "%")

print("---- 1 Month ----")
print("Avg Return (All):", round(one_stats[0], 2), "%")
print("% NonProfitable:", round(one_stats[1], 2), "%")
print("Avg Loss (NonProfitable Only):", round(one_stats[2], 2), "%")

print("\n---- 3 Months ----")
print("Avg Return (All):", round(three_stats[0], 2), "%")
print("% NonProfitable:", round(three_stats[1], 2), "%")
print("Avg Loss (NonProfitable Only):", round(three_stats[2], 2), "%")

print("\n---- 6 Months ----")
print("Avg Return (All):", round(six_stats[0], 2), "%")
print("% NonProfitable:", round(six_stats[1], 2), "%")
print("Avg Loss (NonProfitable Only):", round(six_stats[2], 2), "%")
```
