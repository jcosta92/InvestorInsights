```python
ticker = "MU"  # << set your ticker here
```

```python
# --- Imports ---
import re
import datetime as dt
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pathlib import Path

# --- Constants ---
base = f"https://stockanalysis.com/stocks/{ticker.lower()}/financials"
pages = [
    base + "/",                       # Overview (Income Statement summary)
    base + "/balance-sheet/",         # Balance Sheet
    base + "/cash-flow-statement/",   # Cash Flow
    base + "/ratios/",                # Ratios
]

headers_req = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}

# Output path
today = dt.date.today().isoformat()  # YYYY-MM-DD
out_dir = Path(r"C:\Users\JosGon\Dropbox\Dinheiro\Investimentos\01 Curso de Investimentos\07 Accoes\01 Teses de Investimento\02 GO\00 Helper")
out_dir.mkdir(parents=True, exist_ok=True)
out_path = out_dir / f"{ticker.upper()}_{today}.xlsx"

# --- Helpers (inline, not user-defined functions) ---
# We’ll parse the head row <tr> with <th>, skipping “Fiscal Year” and “2016 - 2019”
headers_full = None
dfs = []

for url in pages:
    html = requests.get(url, headers=headers_req, timeout=30).text
    soup = BeautifulSoup(html, "html.parser")

    # Find the main table: look for a header row that has the first cell "Fiscal Year"
    head_tr = None
    for tr in soup.find_all("tr"):
        ths = tr.find_all("th")
        if not ths:
            continue
        texts = [th.get_text(strip=True) for th in ths]
        if any("Fiscal Year" in t for t in texts):
            head_tr = tr
            break

    if head_tr is None:
        raise RuntimeError(f"Could not find the header row on: {url}")

    # Build header labels
    cols = []
    for th in head_tr.find_all("th"):
        text = th.get_text(strip=True)
        if text in ("Fiscal Year", "2016 - 2019"):
            continue
        # Normalize "FY 2024" -> "FY24", keep "Current"/"TTM" as-is for now
        m = re.match(r"FY\s*(\d{4})", text, flags=re.I)
        if m:
            yr = m.group(1)
            cols.append(f"FY{yr[-2:]}")  # FY24 style
        else:
            cols.append(text)  # "Current" or "TTM"
    headers_full = cols  # keep last seen; they should be the same pattern across pages

    # Parse rows from tbody (metric + values)
    page_rows = []
    idx_names = []

    # The body has <tr> with first <td> = metric name; subsequent <td> = values
    for tr in soup.find_all("tr"):
        tds = tr.find_all("td")
        if not tds:
            continue

        # Metric name from first td (prefer link text, else the visible text)
        metric_name = None
        a = tds[0].find("a")
        if a and a.get_text(strip=True):
            metric_name = a.get_text(strip=True)
        else:
            # sometimes nested <span>/<div> carry the label
            label = tds[0].get_text(" ", strip=True)
            # clean excessive spaces / icons
            metric_name = re.sub(r"\s+", " ", label).strip()

        # Collect numeric cells, ignoring “Upgrade” cells
        vals = []
        for td in tds[1:]:
            txt = td.get_text(strip=True)
            if not txt or txt.lower().startswith("upgrade"):
                continue
            vals.append(txt)

        # Keep as many as headers
        vals = vals[:len(headers_full)]
        # Pad if needed
        if len(vals) < len(headers_full):
            vals += [None] * (len(headers_full) - len(vals))

        idx_names.append(metric_name)
        page_rows.append(vals)

    page_df = pd.DataFrame(page_rows, columns=headers_full, index=idx_names)

    # --- Unify "Current" and "TTM" into a single first column named "Current" ---
    cols_set = page_df.columns.tolist()
    has_current = "Current" in cols_set
    has_ttm = "TTM" in cols_set

    if has_current and has_ttm:
        # combine: prefer Current where present else TTM
        combined = page_df["Current"].where(page_df["Current"].notna(), page_df["TTM"])
        page_df = page_df.drop(columns=["TTM"])
        page_df.insert(0, "Current", combined)
    elif (not has_current) and has_ttm:
        # rename TTM -> Current and move to first col
        page_df = page_df.rename(columns={"TTM": "Current"})
        # Move to front
        new_cols = ["Current"] + [c for c in page_df.columns if c != "Current"]
        page_df = page_df[new_cols]
    else:
        # Ensure "Current" is first if it exists
        if has_current:
            new_cols = ["Current"] + [c for c in page_df.columns if c != "Current"]
            page_df = page_df[new_cols]

    # --- Replace dot with comma in all string cells (values only) ---
    # Keep headers unchanged; values become strings with commas as decimal separators.
    page_df = page_df.map(lambda x: x.replace(",", "") if isinstance(x, str) else x)
    page_df = page_df.map(lambda x: x.replace(".", ",") if isinstance(x, str) else x)

    dfs.append(page_df)

# --- Combine all pages vertically (append all rows) ---
combined = pd.concat(dfs, axis=0, sort=False)

# Optional: drop duplicate row names if any (keep first occurrence)
combined = combined[~combined.index.duplicated(keep="first")]

# --- Reorder columns: Current first, then FY24, FY23, FY22, FY21, FY20, TTM/other if present ---
preferred_order = ["Current", "FY24", "FY23", "FY22", "FY21", "FY20"]
ordered = [c for c in preferred_order if c in combined.columns]
rest = [c for c in combined.columns if c not in ordered]
combined = combined[ordered + rest]

```

```python
# --- Save to Excel ---
with pd.ExcelWriter(out_path, engine="xlsxwriter") as writer:
    combined.to_excel(writer, sheet_name="All Pages")

print(f"Saved: {out_path}")
```
