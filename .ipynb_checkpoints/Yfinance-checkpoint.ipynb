```python
def compute_cagr_from_history(hist: pd.DataFrame):
    """
    Compute CAGR using Adj Close if available, else fallback to Close.
    """
    if hist is None or hist.empty:
        return None
    
    # choose the correct price column
    price_column = "Adj Close" if "Adj Close" in hist.columns else (
                    "Close" if "Close" in hist.columns else None)
    
    if price_column is None:
        return None
    
    hist = hist.dropna(subset=[price_column])
    if hist.empty:
        return None
    
    start_price = hist[price_column].iloc[0]
    end_price   = hist[price_column].iloc[-1]
    
    if start_price <= 0 or end_price <= 0:
        return None
    
    start_date = hist.index[0]
    end_date   = hist.index[-1]
    years = (end_date - start_date).days / 365.25
    
    if years <= 0:
        return None
    
    return (end_price / start_price) ** (1 / years) - 1
```

```python
def fetch_stock_snapshot(ticker: str) -> dict:
    """
    Fetch maximum useful fields from yfinance for a given ticker.
    Returns a dict you can convert to DataFrame or JSON.
    """
    tk = yf.Ticker(ticker)
    
    # In newer yfinance versions .info may be deprecated, use get_info()
    try:
        info = tk.info
    except Exception:
        info = tk.get_info()
    
    # Basic current price / fast info
    fast = getattr(tk, "fast_info", {}) or {}
    
    # Historical data for performance & CAGR
    hist_6m = tk.history(period="6mo")
    hist_5y = tk.history(period="5y")
    hist_10y = tk.history(period="10y")
    
    cagr_5y = compute_cagr_from_history(hist_5y)
    cagr_10y = compute_cagr_from_history(hist_10y)
    
    # Analyst & recommendations
    # These are DataFrames; we’ll just grab some key aggregate fields from info
    analysis = None
    try:
        analysis = tk.analysis
    except Exception:
        pass
    
    # Build result dict
    data = {
        "ticker": ticker,
        
        # --- Company profile-ish ---
        "country": info.get("country"),
        "industry": info.get("industry"),
        "sector": info.get("sector"),
        "employees_full_time": info.get("fullTimeEmployees"),
        "website": info.get("website"),
        "long_business_summary": info.get("longBusinessSummary"),
        
        # --- Market data & valuation ---
        "current_price": info.get("currentPrice") or fast.get("lastPrice") or fast.get("last_price"),
        "currency": info.get("currency") or fast.get("currency"),
        "market_cap": info.get("marketCap"),
        "trailing_PE": info.get("trailingPE"),
        "forward_PE": info.get("forwardPE"),
        "price_to_book": info.get("priceToBook"),
        "price_to_sales_TTM": info.get("priceToSalesTrailing12Months"),
        "peg_ratio": info.get("pegRatio"),
        
        # --- Dividend / income ---
        "dividend_yield": info.get("dividendYield"),          # as decimal (e.g. 0.015)
        "dividend_rate": info.get("dividendRate"),
        "fiveYear_avg_dividend_yield": info.get("fiveYearAvgDividendYield"),
        "payout_ratio": info.get("payoutRatio"),
        
        # --- 52-week range ---
        "52_week_low": info.get("fiftyTwoWeekLow"),
        "52_week_high": info.get("fiftyTwoWeekHigh"),
        
        # --- Analyst summary from Yahoo ---
        "target_mean_price": info.get("targetMeanPrice"),
        "target_median_price": info.get("targetMedianPrice"),
        "target_high_price": info.get("targetHighPrice"),
        "target_low_price": info.get("targetLowPrice"),
        "number_of_analyst_opinions": info.get("numberOfAnalystOpinions"),
        "recommendation_key": info.get("recommendationKey"),   # e.g. 'buy', 'hold'
        "recommendation_mean": info.get("recommendationMean"), # numeric score
        
        # --- CAGR from price history ---
        "cagr_5y": cagr_5y,
        "cagr_10y": cagr_10y,
        
        # --- Additional fields that might be useful later ---
        "beta": info.get("beta"),
        "enterprise_value": info.get("enterpriseValue"),
        "enterpriseToRevenue": info.get("enterpriseToRevenue"),
        "enterpriseToEbitda": info.get("enterpriseToEbitda"),
        "profit_margins": info.get("profitMargins"),
        "operating_margins": info.get("operatingMargins"),
        "gross_margins": info.get("grossMargins"),
        "ebitda_margins": info.get("ebitdaMargins"),
        "return_on_equity": info.get("returnOnEquity"),   # if available
        "return_on_assets": info.get("returnOnAssets"),   # if available
        
        # Raw data lengths so you know if 5Y/10Y are full or truncated
        "rows_history_6m": len(hist_6m) if hist_6m is not None else 0,
        "rows_history_5y": len(hist_5y) if hist_5y is not None else 0,
        "rows_history_10y": len(hist_10y) if hist_10y is not None else 0,
    }
    
    return data
```

```python
def plot_price_history(ticker: str, period: str = "5y"):
    tk = yf.Ticker(ticker)
    hist = tk.history(period=period)

    if hist is None or hist.empty:
        print(f"No historical data for {ticker} with period={period}")
        return
    
    # Decide which column to plot
    if "Adj Close" in hist.columns:
        price_col = "Adj Close"
    elif "Close" in hist.columns:
        price_col = "Close"
    else:
        print(f"No 'Adj Close' or 'Close' column found for {ticker}. Columns: {hist.columns.tolist()}")
        return

    plt.figure(figsize=(10, 4))
    plt.plot(hist.index, hist[price_col])
    plt.title(f"{ticker} - {price_col} ({period})")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# 5 years
plot_price_history(ticker, period="5y")

# 6 months
plot_price_history(ticker, period="6mo")
```
