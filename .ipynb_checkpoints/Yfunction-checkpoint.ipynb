```python
"""
yfinance_data.py

Módulo para buscar dados de um ticker via yfinance e gerar gráficos em PNG.

Funções principais:
- get_yf_data(mainticker, period, interval)
    -> df_history, info, recs_summary
- save_yf_charts(df_history, mainticker, out_dir)
    -> dict com paths dos ficheiros PNG
- get_yf_package(mainticker, period, interval, charts_dir)
    -> dicionário com tudo (dados + paths dos gráficos)
"""

from pathlib import Path
from typing import Dict, Any, Tuple, Optional

import yfinance as yf
import matplotlib
matplotlib.use("Agg")  # backend sem janela gráfica (bom para servidores / scripts)
import matplotlib.pyplot as plt
import pandas as pd


def get_yf_data(
    mainticker: str,
    period: str = "5y",
    interval: str = "1d",
) -> Tuple[pd.DataFrame, Dict[str, Any], Optional[pd.DataFrame]]:
    """
    Faz o download dos dados do yfinance para um ticker.

    Parameters
    ----------
    mainticker : str
        Símbolo do ticker (ex: "AAPL", "MSFT").
    period : str, opcional
        Período do histórico (ex: "1y", "5y", "max").
    interval : str, opcional
        Intervalo do histórico (ex: "1d", "1wk", "1mo").

    Returns
    -------
    df_history : pd.DataFrame
        Histórico de preços com colunas típicas (Open, High, Low, Close, Volume...).
    info : dict
        Dicionário com as informações do ticker (empresa, setor, etc.).
    recs_summary : pd.DataFrame ou None
        Resultado de tk.get_recommendations_summary(), se existir; caso contrário, None.
    """
    tk = yf.Ticker(mainticker)

    # Histórico de preços
    df_history = tk.history(period=period, interval=interval)

    # Info da empresa
    info = tk.info  # normalmente é um dict grande

    # Resumo de recomendações (nem todas as versões têm este método)
    recs_summary = None
    try:
        if hasattr(tk, "get_recommendations_summary"):
            recs_summary = tk.get_recommendations_summary()
        elif hasattr(tk, "recommendations_summary"):
            # fallback se for atributo em vez de método
            recs_summary = tk.recommendations_summary
    except Exception as e:
        # Se der erro, simplesmente deixamos como None
        print(f"[WARN] Não foi possível obter recommendations_summary para {mainticker}: {e}")

    return df_history, info, recs_summary


def save_yf_charts(
    df_history: pd.DataFrame,
    mainticker: str,
    out_dir: str = "charts",
) -> Dict[str, Path]:
    """
    Cria e guarda gráficos em ficheiros PNG para um dado df_history.

    Gráficos criados:
    - price: evolução do preço de fecho
    - volume: barras de volume

    Parameters
    ----------
    df_history : pd.DataFrame
        DataFrame com histórico de preços (resultado de get_yf_data()).
    mainticker : str
        Símbolo do ticker, usado nos títulos e nomes dos ficheiros.
    out_dir : str, opcional
        Diretório onde os PNG serão guardados.

    Returns
    -------
    chart_paths : dict
        Dicionário com os paths dos ficheiros PNG, por exemplo:
        {
            "price": Path("charts/AAPL_price.png"),
            "volume": Path("charts/AAPL_volume.png")
        }
    """
    out_path = Path(out_dir)
    out_path.mkdir(parents=True, exist_ok=True)

    chart_paths: Dict[str, Path] = {}

    # --- Gráfico de preço de fecho ---
    if "Close" in df_history.columns:
        price_png = out_path / f"{mainticker}_price.png"

        fig_price, ax_price = plt.subplots()
        ax_price.plot(df_history.index, df_history["Close"])
        ax_price.set_title(f"{mainticker} - Closing Price")
        ax_price.set_xlabel("Date")
        ax_price.set_ylabel("Price")

        fig_price.tight_layout()
        fig_price.savefig(price_png, bbox_inches="tight")
        plt.close(fig_price)

        chart_paths["price"] = price_png
    else:
        print(f"[WARN] df_history não tem coluna 'Close' para {mainticker}, não foi gerado gráfico de preço.")

    # --- Gráfico de volume ---
    if "Volume" in df_history.columns:
        volume_png = out_path / f"{mainticker}_volume.png"

        fig_vol, ax_vol = plt.subplots()
        ax_vol.bar(df_history.index, df_history["Volume"])
        ax_vol.set_title(f"{mainticker} - Volume")
        ax_vol.set_xlabel("Date")
        ax_vol.set_ylabel("Volume")

        fig_vol.tight_layout()
        fig_vol.savefig(volume_png, bbox_inches="tight")
        plt.close(fig_vol)

        chart_paths["volume"] = volume_png
    else:
        print(f"[WARN] df_history não tem coluna 'Volume' para {mainticker}, não foi gerado gráfico de volume.")

    return chart_paths


def get_yf_package(
    mainticker: str,
    period: str = "5y",
    interval: str = "1d",
    charts_dir: str = "charts",
) -> Dict[str, Any]:
    """
    Função "all-in-one" para integrar no teu pipeline.

    Parameters
    ----------
    mainticker : str
        Símbolo do ticker (ex: "AAPL").
    period : str, opcional
        Período do histórico.
    interval : str, opcional
        Intervalo do histórico.
    charts_dir : str, opcional
        Diretório onde os gráficos serão guardados.

    Returns
    -------
    result : dict
        {
            "ticker": mainticker,
            "history": df_history,
            "info": info,
            "recs_summary": recs_summary,
            "charts": {
                "price": Path(...),
                "volume": Path(...)
            }
        }
    """
    df_history, info, recs_summary = get_yf_data(
        mainticker=mainticker,
        period=period,
        interval=interval,
    )

    chart_paths = save_yf_charts(
        df_history=df_history,
        mainticker=mainticker,
        out_dir=charts_dir,
    )

    result: Dict[str, Any] = {
        "ticker": mainticker,
        "history": df_history,
        "info": info,
        "recs_summary": recs_summary,
        "charts": chart_paths,
    }
    return result


if __name__ == "__main__":
    # Pequeno teste rápido se correres o ficheiro diretamente:
    test_ticker = "AAPL"
    print(f"Teste rápido com {test_ticker}...")
    pkg = get_yf_package(test_ticker, period="1y", interval="1d", charts_dir="charts_test")
    print("Histórico:", pkg["history"].tail(3))
    print("Info keys:", list(pkg["info"].keys())[:10])
    print("Recs summary:", pkg["recs_summary"])
    print("Charts:", pkg["charts"])
```
